## Merging details db

```python
import re
detail_df = pd.read_csv('game_details.csv')
detail_df = detail_df.set_index('title')
detail_df['category'] = [re.findall('\w+',x) for x in detail_df['category']]
detail_df['subcategory'] = [x.replace("'","").strip(" []").split(", ") for x in detail_df['subcategory']]
```

```python
# test = detail_df['subcategory'][1]
# test = test.strip(" []")
# test = test.replace("'","")
# test = test.split(", ")
# type(test)
```

```python
db = pd.merge(boardgame_df,detail_df,how="inner",on="title")
db.describe()
```

```python
db['avg_time']=(db['min_time']+db['max_time'])/2
db.head()
```

```python
plt.figure(figsize=(12,6))
plt.scatter(db['age'], db['complexity'])
plt.ylabel('complexity')
plt.xlabel('age')
```

```python
plt.figure(figsize=(12,6))
db.plot(kind='scatter',x='avg_rating', y='complexity', figsize=(12,6))
plt.xlabel('avg_rating')
plt.ylabel('complexity')
```

```python
db_category = db[['avg_rating','num_rating','year','age','complexity','min_player','max_player','avg_time','category']]
db_category = db_category[db_category['category'].map(lambda c: len(c)) > 0]
db_category = db_category.reset_index()
db_category
```

```python
index=0
list_ = []
for item in db_category.category:
    list_.extend(map(lambda x: [index, x], item))
    index += 1
categories = pd.DataFrame(list_, columns=['index', 'category'])
categories['category'].unique()
```

```python
index=0
list_ = []
for item in db.subcategory:
    list_.extend(map(lambda x: [index, x], item))
    index += 1
subcategories = pd.DataFrame(list_, columns=['index', 'subcategory'])
subcategories['subcategory'].unique()
```

```python
#subcategories['subcategory'].hist(figsize=(12,4))
subCounts = subcategories['subcategory'].value_counts()
subCounts.head(20)
```

```python
db_category = pd.merge(db_category.drop('category', axis=1), categories, how='outer', left_index=True, right_on='index').set_index('title')
db_category = db_category[db_category.category != "s"]
db_category = db_category.drop('index', 1)
db_category
```
