```python
%matplotlib inline
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
plt.style.use('ggplot')
```

```python
df = pd.read_csv('boardgames.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df=df.set_index('title')
df.describe()
```

```python
df['year'].value_counts()
```

```python
plt.hist(df['year'])
```

```python
df.sort_values('year')
```

```python
boardgame_df = df.loc[df.year>1950,:]
boardgame_df = boardgame_df.loc[boardgame_df.year<2100,:]
boardgame_df = boardgame_df.loc[boardgame_df.num_rating>100,:]
boardgame_df
```

```python
boardgame_df['year'].value_counts()
```

```python
plt.figure(figsize=(20,6))
plt.hist(boardgame_df['year'],bins=70)
```

```python
legacy = df.loc[df.year<1950,:]
legacy = legacy.loc[legacy.num_rating>100]
legacy.sort_values('num_rating',ascending=False).head(10) #top 10 legacy games by popularity
```

```python
plt.figure(figsize=(20,6))
plt.scatter(legacy['num_rating'], legacy['avg_rating'])
```

```python
plt.figure(figsize=(20,10))
plt.scatter(boardgame_df['year'], boardgame_df['avg_rating'])
```

```python
year_df = boardgame_df[['year', 'avg_rating']]
plt.figure(figsize=(20,12))
year_df.boxplot(by='year', column='avg_rating')
```

```python
boardgame_df['num_rating'].mean()
```
